```python
import pandas as pd

df = pd.read_csv(r"../data/cleaned/online_retail_cleaned.csv",parse_dates=["InvoiceDate"])

df.shape
```

```python
df.head()
```

```python
snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)
snapshot_date
```

```python
customer_df = df.groupby("CustomerID").agg(Recency=("InvoiceDate", lambda x: (snapshot_date - x.max()).days),Frequency=("InvoiceNo", "nunique"),Monetary=("Revenue", "sum")).reset_index()

customer_df.shape
```

```python
customer_df.head()
```

```python
customer_df.describe()
```

```python
customer_df.sort_values("Monetary", ascending=False).head()
```

```python
customer_df.to_csv(r"../data/cleaned/customer_level_rfm_base.csv",index=False)
```
